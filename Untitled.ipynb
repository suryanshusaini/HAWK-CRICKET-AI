```python
import pandas as pd

# Load your new master dataset
df = pd.read_csv('all_matches_deliveries.csv')

# Display the first 5 rows to see what it looks like
print("Dataset loaded successfully!")
df.head()
```

```python
# Get a summary of the data and check for missing values
df.info()

print("\nMissing values per column:")
df.isnull().sum()
```

```python
# Create separate 'over_number' and 'ball_in_over' columns
df['over'] = df['over'].astype(str)
df['over_number'] = df['over'].apply(lambda x: int(x.split('.')[0]))
df['ball_in_over'] = df['over'].apply(lambda x: int(x.split('.')[1]))

# Create the 'bowling_team' column
teams_per_match = df.groupby('match_id')['batting_team'].unique()

def get_bowling_team(row):
    match_teams = teams_per_match.loc[row['match_id']]
    # Handle cases where a match might have less than 2 teams recorded
    if len(match_teams) < 2:
        return "Unknown"
    return match_teams[0] if row['batting_team'] == match_teams[1] else match_teams[1]

df['bowling_team'] = df.apply(get_bowling_team, axis=1)

# Create the 'match_phase' column for ODIs
def get_match_phase(over):
    if over <= 10:
        return 'Powerplay'
    elif over <= 40:
        return 'Middle'
    else:
        return 'Death'

df['match_phase'] = df['over_number'].apply(get_match_phase)

print("Feature engineering complete! Here are the new columns:")
df[['batting_team', 'bowling_team', 'over_number', 'ball_in_over', 'match_phase']].head()
```
